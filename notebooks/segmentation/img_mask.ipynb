```python
sample_path = "/content/drive/MyDrive/Colab Notebooks/Pulmonary Fibrosis/results/results_sample_data.csv"
sample_df = pd.read_csv(sample_path,index_col = [0])
sample_df
```

```python
unique_id = pd.unique(sample_df.Patient_id)
unique_id
```

```python
def class_images(img_count, img_name):
  img_ = float(img_name.split(".")[0])
  per_20 = float(img_count*0.30)
  per_45 = float(img_count*0.475)
  per_55 = float(img_count*0.525)
  per_75 = float(img_count*0.70)
  if img_ <= per_20:
    return 1
  elif img_ <= per_45:
    return 2
  elif img_ <= per_55:
    return 3
  elif img_ <= per_75:
    return 4
  else:
    return 5
```

```python
class_images(300,"51.png")
```

```python
all_data = list()

for patient_id in os.listdir(masks_path):
  if patient_id in unique_id:
    sorted_files = sorted(os.listdir(masks_path + patient_id), key=lambda x: int(x.split('.')[0]))
    num_files = len(sorted_files)
    for img_name in sorted_files:
      counter = list()
      counter.append(patient_id)
      counter.append(imgs_path + patient_id + "/" + img_name)
      counter.append(masks_path + patient_id + "/" + img_name)
      counter.append(targets_path + patient_id + "/" + img_name)
      counter.append(class_images(num_files , img_name))
      all_data.append(counter)

```

```python
len(all_data)
```

```python
df = pd.DataFrame(all_data, columns=['Patient_id',"Image_path","Mask_path","Image_mask_path","Class"])
df
```
