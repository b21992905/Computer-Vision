```python
def show_dcom_file(path):

  # Load the DICOM file
  ds = pydicom.dcmread(path)

  # Access DICOM metadata
  print("Patient ID:", ds.PatientID)
  print("Modality:", ds.Modality)
  # print("Study Description:", ds.StudyDescription)
  print("Series Description:", ds.SeriesDescription)

  # Access pixel data (assuming it's an image)
  pixel_data = ds.pixel_array

  # Display the image (optional)

  plt.imshow(pixel_data, cmap=plt.cm.gray)
  plt.show()
```

```python
show_dcom_file(folder_path + patient + dcm_file)
```

# **talisman**

```python
import os
labels={"emphysema":0,"micronodules":0,"fibrosis":0,"ground_glass":0,"healthy":0}

def filtrele(dizin_yolu):
    hastalar = {}
    global labels
    for dosya in os.listdir(dizin_yolu):
        dosya_adi, uzanti = os.path.splitext(dosya)
        if uzanti.lower() == '.tif':
            i=3
            #print(dosya_adi[0])
            if dosya_adi.lower().startswith("em"):
                hastalik = "emphysema"
            elif dosya_adi.lower().startswith("mi"):
                hastalik = "micronodules"
            elif dosya_adi.lower().startswith("fi"):
                hastalik = "fibrosis"
            elif dosya_adi.lower().startswith("gr"):
                hastalik = "ground_glass"
            else:
                hastalik = "healthy"

                i=4
            labels[hastalik]+=1

            if "patient" in dosya_adi.lower():
                hasta_numarasi = dosya_adi.split("patient")[1][0:]
            else:
                print("mall")
                hasta_numarasi = dosya_adi[7:]
            #print(hasta_numarasi)

            if not hasta_numarasi in hastalar:
              hastalar[hasta_numarasi]={hastalik}
            else:
              hastalar[hasta_numarasi].add(hastalik)


    return hastalar

dosya_dizini = "/content/drive/MyDrive/HUG-data/ILD_DB_talismanTestSuite"  # Dosyaların bulunduğu dizin
hastalar = filtrele(dosya_dizini)

# Elde edilen hastalar listesini yazdırma
for hasta in hastalar:
    print(f"Hasta Numarası: {hasta}, Hastalık: {hastalar[hasta]}")
```
